```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

df = pd.read_csv('TED_Preprocessed.csv')
df = df[df['views'] < 100000]
```

```python
del df['video_link']
del df['date_month_year']
```

### Finding Correlations from scratch

### a) Views-to-comments

```python
factors = []
for i in df['comments']/df['views']:
    factors.append(round(i,5))
    
df['views_to_comments'] = factors
```

```python
df.head(2)
```

```python
print('Max = ', max(df['views_to_comments']))
print('Min = ', min(df['views_to_comments']))
print('Avg = ', (df['views_to_comments'].mode()))
```

#### b) Adding predicted comments column

```python
factor = df['views_to_comments'].mode()[0]
factor
```

```python
predicted_comments = []

for i in df['views']:
    predicted_comments.append(int(i*factor))
    
df['predicted_comments'] = predicted_comments
```

```python
df.head(2)
```

#### >> Correlation between [Likes, Predicted Likes]

```python
data = []
for i in df.values:
    data.append([i[2], i[4], i[10]])
df_ = pd.DataFrame(data, columns=['views', 'comments', 'predicted_comments'])
```

```python
views_ = df_.sort_values(by='views')['views']
comments_ = df_.sort_values(by='views')['comments']
predicted_comments = df_.sort_values(by='views')['predicted_comments']
```

```python
fig, ax = plt.subplots(figsize=(15,4))
ax.plot(views_, comments_, label='Comments')
ax.plot(views_, predicted_comments, label = 'Predicted comments')
plt.xlabel('Views')
plt.ylabel('Number of comments')
plt.title('Comments and Predicted likes comparison')
plt.legend()
plt.show()
```

### Finding Loss using Mean error

```python
total_error = []
for i in df.values:
    t = i[4] - i[10]
    if t>=0 :
        total_error.append(t)
    else:
        total_error.append(-t)
        
avg_error = sum(total_error)/len(total_error)
avg_error
```

#### Finding view to comment ratio

```python
start = int(df['views_to_comments'].min() * 100000)
end = int(df['views_to_comments'].max() * 100000) + 1
factors = []
for i in range(start, end):
    factors.append(i/100000)
```

### Combining Factor + Error + Ratios

```python
comments = np.array(df['comments'])
data = []

for i in tqdm(range(start, end)):
    factor = i/100000
    
    pred_comments = []
    for j in df['views']:                       # Predicting comments for each factor
        pred_comments.append(int(j*factor))
    pred_comments = np.array(pred_comments)
        
    total_error = []
    for k in range(len(comments)):
        t = comments[k] - pred_comments[k]
        if t>=0 :
            total_error.append(t)
        else:
            total_error.append(-t)
    total_error = np.array(total_error)  
    
    error = int(total_error.mean())
    
    data.append([factor, error])
```

```python
df_ = pd.DataFrame(data, columns=['factor', 'error'])
df_.sort_values(by='error').head(1)
```

```python
count = 0
for i in df_.values:
    if i[1] == 648:
        count += 1
```

```python
df_ = pd.DataFrame(data, columns=['factor', 'error'])
final_factor = df_.sort_values(by='error').head(count).mean()['factor']
final_factor = round(final_factor, 5)
final_factor
```

```python
comments_ = []
for i in df['views']:
    comments_.append(int(i * final_factor))
    
df['predicted_comments'] = comments_
df.head()
```

### Actutal to predicted likes with best fit factor

```python
data = []
for i in df.values:
    data.append([i[2], i[4], i[10]])
df_ = pd.DataFrame(data, columns=['views', 'comments', 'predicted_comments'])

views_ = df_.sort_values(by='views')['views']
comments_ = df_.sort_values(by='views')['comments']
predicted_comments = df_.sort_values(by='views')['predicted_comments']

fig, ax = plt.subplots(figsize=(15,4))
ax.plot(views_, comments_, label='Comments')
ax.plot(views_, predicted_comments, label = 'Predicted comments')
plt.xlabel('Views')
plt.ylabel('Number of comments')
plt.title('Comments and Predicted likes comparison')
plt.legend()
plt.show()
```
